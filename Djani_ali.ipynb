### Data processing of different historical portfolios
#### Difficulty:
- We don't know how many sheet the excel file contains
- Different mandats have different historical dates
- Different mandats have a different ISIN universe


```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None) # to be able to display all columns of a dataframe
```

```python
# Cells that are in the same column as ISIN, but are not ISIN

delete_list = ['CHF',  'EUR', 'GBP', 'Autre', 'USD', 'Autres', 'TOTAL', 'Commodities', 'Obligations',
              'NON EUR', 'CAD', 'AUD', 'JPY', 'EMCUR', 'Ass. pf action', 'PAC_XJPY',
              'Courbe CHF - duration', 'Courbe EUR - duration', 'Actions Monde',
              'Actions EM', 'Actions EMU', 'Obligations EM', 'Obligations High Yield',
              'Obligations Monde', 'Obligations EUR', 'Liquidités', 'High Yield', 
              'Actions', 'Obligation', 'Gestion décorrélée', 'Gest. asym. action',
              'Courbe USD - duration', 'Courbe Monde - duration', 'Hedge Funds', 'TOTAL',
              '% Gouvernementales - courbe CHF', '% Corporate - courbe CHF',
              '% Gouvernementales - courbe EUR', '% Corporate - courbe EUR',
              '% Gouvernementales - courbe USD', '% Corporate - courbe USD',
              '% Gouvernementales - courbe Monde', '% Corporate - courbe Monde',
              '% Gouvernementales - Courbe CHF', '% Corporate - Courbe CHF',
              '% Gouvernementales - Courbe EUR', '% Corporate - Courbe EUR',
              '% Gouvernementales - Courbe USD', '% Corporate - Courbe USD',
              '% Gouvernementales - Courbe Monde', '% Corporate - Courbe Monde']
```

```python
df_excel = pd.read_excel('ss_histo.xlsx', parse_dates=False, sheet_name=None, header=3) # sheet_name=None => input all excel sheets in a dataframe dictionary
```

```python
i = 1
df = list(df_excel.values())[i] # values => dataframe
sheet_name = list(df_excel.keys())[i] # keys => name of the sheet/mandat
print('Mandat :', sheet_name)
df.head()
```

```python
df = df.dropna(how='all')
df = df[-df['No ISIN'].isin(delete_list)] # Eliminate rows whose ISIN is in the delete list
df.head()
```

```python
df = df.drop(df.columns[[0,1,3]], axis=1)
df.head()
```

```python
list_date = list(df.columns)[1:] # Get all dates available of a particular sheet
# Pivot the dataframe in a more robust architecture
df = df.melt(id_vars='No ISIN', value_vars=list_date, var_name='histo_date', value_name='weight')
df = df[df['No ISIN'].notna()] # drop rows when ISIN is NaN
df = df.dropna(subset=['weight'])
df = df[df['weight'] != 0] # remove weight of 0% to get only valuable data
df.head()
```

```python
df['portfolio'] = 'Star Select'
df['strategie'] = sheet_name
df.head()
```

```python
# string processing
if 'III' in sheet_name:
    df['on/off'] = 'off'
    strategie = sheet_name.replace(' - III', '') # remove a substring
    strategie = strategie.replace('_III', '')
    df['strategie'] = strategie
else:
    df['on/off'] = 'on'
    df['strategie'] = sheet_name
    
df.head()
```

```python
nbr_sheet = len(list(df_excel.values()))

# Iterate through all sheet of the excel file
for i in range(nbr_sheet):
    
    df = list(df_excel.values())[i] # values => dataframe
    sheet_name = list(df_excel.keys())[i] # keys => name of the sheet/mandat
    if 'No ISIN' not in df.columns: # Ignore a sheet if does not contain ISIN column
        continue
        
    df = df.dropna(how='all') # drop row when all columns are NaN
    df = df[-df['No ISIN'].isin(delete_list)] # Eliminate rows whose ISIN is in the delete list
    
    df = df.drop(df.columns[[0,1,3]], axis=1) # drop unused columns
    list_date = list(df.columns)[1:] # Get all dates available of a particular sheet
    
    # Pivot the dataframe in a more robust architecture
    df = df.melt(id_vars='No ISIN', value_vars=list_date, var_name='histo_date', value_name='weight')
    df = df[df['No ISIN'].notna()] # drop rows when ISIN is NaN
    df = df.dropna(subset=['weight'])
    df = df[df['weight'] != 0] # remove weight of 0% to get only valuable data
    
    df['portfolio'] = 'Star Select'
    df['strategie'] = sheet_name
    
    # string processing
    if 'III' in sheet_name:
        df['on/off'] = 'off'
        strategie = sheet_name.replace(' - III', '') # remove a substring
        strategie = strategie.replace('_III', '')
        df['strategie'] = strategie
    else:
        df['on/off'] = 'on'
        df['strategie'] = sheet_name
    
    df['weight'] = df['weight'].astype('float64') # change type of column
    df['weight'] = df['weight']
    
    # At first iteration initialize the final dataframe
    if i==0:
        df_port = df
    # Concatenate all portfolios in the final dataframe
    else:
        df_port = pd.concat([df_port, df], axis=0)


df_port['histo_date'] = df_port['histo_date'].dt.date
df_port = df_port.drop_duplicates()
```
